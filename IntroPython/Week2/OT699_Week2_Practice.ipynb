### Reading and Writing CSV Files

```python
filename = "week2.csv"
```

```python
pd.read_csv(filename)
```

```python
# Read it in as a data ina  variable - Here I call the dataframe as mydf
mydf = pd.read_csv(filename)
mydf.head()
```

#### Manipulating Data

```python
# Let's look at the first X rows only
mydf.head(5)
```

```python
# Get average age of this dataset
mydf['Age'].mean()
mydf['Age'].mean().round(2)
```

```python
# Convert weight from pounds to kilograms (1 pound is 0.453596 Kgs) in a new column
p2kg_rate = 0.453596
mydf['Weight(Kgs)'] = mydf['Weight'] * p2kg_rate
mydf.sample(5)
```

```python
# Save the new dataframe to a new CSV file
mydf.to_csv("week2_updated20260716.csv")
```

#### Assignment - Write a function that would tell you (true or false) if the number you input is greater than or equal to 10 or not

```python
def comp(x):
    if x >= 10:
        print("TRUE!")
    else:
        print("FALSE!")
```

```python
comp(5)
```

```python
comp(18)
```

```python
for n in range (0,21):
    print("For the entry", n, ", the verdict is:")
    comp(n)
```
